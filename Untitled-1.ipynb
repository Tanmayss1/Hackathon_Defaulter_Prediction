### Loan Purpose vs. Loan Status:

"Debt Consolidation" is the most common loan purpose for both paid and defaulted loans.
"Car" and "Educational" purposes have a significantly higher percentage of defaulted loans compared to paid loans.
"Credit Card," "Home Improvement," and "Major Purchase" purposes have a higher percentage of paid loans compared to defaulted loans.

### Employment Length vs. Loan Status:

Borrowers with "10+ years" of employment history have the highest number of both paid and defaulted loans.
Borrowers with "< 1 year" of employment history have a relatively high percentage of defaulted loans compared to paid loans.
Employment length does not show a strong pattern in the data, as the distribution of defaulted and paid loans is relatively consistent across different employment lengths.


### Overall, the majority of loans are paid, but there are variations in loan purposes and employment lengths that may impact the likelihood of default. These insights can help in understanding the factors that contribute to loan status and guide further analysis and modeling.

### Coming back on missing values

```python
df.isnull().sum()
```

```python
#filling with mode values: as these are categorical values
mode_emp_length = df['emp_length'].mode().iloc[0]
mode_emp_title = df['emp_title'].mode().iloc[0]

#filling with median values: as these are numerical values
median_num_actv_bc_tl = df['num_actv_bc_tl'].median()
median_mort_acc = df['mort_acc'].median()
median_tot_cur_bal = df['tot_cur_bal'].median()

```

```python
df['emp_length'].fillna(mode_emp_length, inplace=True)
df['emp_title'].fillna(mode_emp_length, inplace=True)

df['num_actv_bc_tl'].fillna(mean_num_actv_bc_tl, inplace=True)
df['mort_acc'].fillna(median_mort_acc, inplace=True)
df['tot_cur_bal'].fillna(mean_tot_cur_bal, inplace=True)
```

```python
#revmoivng these columns with fewer null values
df.dropna(subset=['pub_rec_bankruptcies'], inplace=True)
df.dropna(subset=['revol_util'], inplace=True)
df.dropna(subset=['title'], inplace=True)
```
